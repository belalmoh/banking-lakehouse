```python
import pandas as pd
import os

DATA_DIR = '/opt/data/sample'

# Verify generated files
for f in os.listdir(DATA_DIR):
    path = os.path.join(DATA_DIR, f)
    size_mb = os.path.getsize(path) / (1024 * 1024)
    print(f'{f}: {size_mb:.2f} MB')
```

```python
# Preview customers
customers = pd.read_csv(f'{DATA_DIR}/customers.csv')
print(f'Customers: {len(customers)} records')
print(f'Segments: {customers["customer_segment"].value_counts().to_dict()}')
print(f'Countries: {customers["country"].value_counts().to_dict()}')
customers.head()
```

```python
# Preview transactions
transactions = pd.read_csv(f'{DATA_DIR}/transactions.csv')
print(f'Transactions: {len(transactions)} records')
print(f'Types: {transactions["transaction_type"].value_counts().to_dict()}')
print(f'Avg amount: {transactions["amount"].mean():.2f}')
print(f'Max amount: {transactions["amount"].max():.2f}')
transactions.head()
```

```python
# Preview AML alerts
aml = pd.read_csv(f'{DATA_DIR}/aml_alerts.csv')
print(f'AML Alerts: {len(aml)} records')
print(f'Types: {aml["alert_type"].value_counts().to_dict()}')
print(f'Severity: {aml["severity"].value_counts().to_dict()}')
aml.head()
```
