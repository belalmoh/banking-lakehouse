# 02 - Bronze Layer Exploration

Explore the raw data ingested into the Bronze Delta Lake layer.
Demonstrates audit trails, schema evolution, and Delta Lake features.

```python
import sys
sys.path.insert(0, '/opt/spark-jobs')

from utils.spark_session import create_spark_session
from utils.delta_helpers import read_delta, get_table_history

spark = create_spark_session(app_name='BronzeExploration')
```

```python
# Read Bronze transactions
transactions = read_delta(spark, 's3a://bronze/transactions')
print(f'Bronze Transactions: {transactions.count()} records')
print(f'Schema:')
transactions.printSchema()
```

```python
# Verify audit columns
audit_cols = ['_ingestion_timestamp', '_source_file', '_batch_id',
              '_data_classification', '_data_residency']
transactions.select(audit_cols).show(5, truncate=False)
```

```python
# Delta Lake version history (time travel)
history = get_table_history(spark, 's3a://bronze/transactions', limit=10)
history.show(truncate=False)
```

```python
# Read Bronze customers
customers = read_delta(spark, 's3a://bronze/customers')
print(f'Bronze Customers: {customers.count()} records')

# Segment distribution
customers.groupBy('customer_segment').count().show()
```
